```python
from PIL import Image, ImageEnhance
from matplotlib import pyplot as plt

kelvin_table = {
    0: (255, 255, 255),
    1000: (255,56,0),
    1500: (255,109,0),
    2000: (255,137,18),
    2500: (255,161,72),
    3000: (255,180,107),
    3500: (255,196,137),
    4000: (255,209,163),
    4500: (255,219,186),
    5000: (255,228,206),
    5500: (255,236,224),
    6000: (255,243,239),
    6500: (255,249,253),
    7000: (245,243,255),
    7500: (235,238,255),
    8000: (227,233,255),
    8500: (220,229,255),
    9000: (214,225,255),
    9500: (208,222,255),
    10000: (204,219,255),}
#     10500: (200, 216, 255),
#     11000: (196, 215, 255), 
#     11500: (193, 212, 255),
#     12000: (191, 211, 255),
#     12500: (188, 209, 255)}


def convert_temp(image, temp):
    r, g, b = kelvin_table[temp]
    matrix = ( r / 255.0, 0.0, 0.0, 0.0,
               0.0, g / 255.0, 0.0, 0.0,
               0.0, 0.0, b / 255.0, 0.0 )
    return image.convert('RGB', matrix)

def plot_image(image):
    plt.figure(figsize=[10, 10])
    plt.imshow(image)
    
def convert_brightness(image, factor):
    img_enhancer = ImageEnhance.Brightness(image)
    return img_enhancer.enhance(factor)

def convert_contrast(image, factor):
    img_enhancer = ImageEnhance.Contrast(image)
    return img_enhancer.enhance(factor)
```

```python
image = Image.open('../data/bright.jpeg')
plot_image(image)
plot_image(convert_temp(image, 0))
plot_image(convert_brightness(image, 0.2))
```

```python
list_ = list(kelvin_table.keys())
f, axarr = plt.subplots(4,5, figsize=(12, 12))
for i in range(4):
    for j in range(5):
        axarr[i, j].imshow(convert_temp(image, list_[i*5 + j]))
```
