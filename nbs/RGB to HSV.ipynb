```python
import cv2
from matplotlib import pyplot as plt
```

```python
def convert_rgb_to_hsv(image_path):
    # Load the image in RGB format
    image = cv2.imread(image_path)
    
    # Convert RGB image to HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    return hsv_image

def plot_hsv(hsv):
#     plt.figure(figsize=[10, 10])
#     plt.imshow(hsv[:, :, 2], cmap='gray')
#     plt.title('v')
#     plt.show()
#     plt.figure(figsize=[10, 10])
#     plt.imshow(hsv[:, :, 1], cmap='gray')
#     plt.title('s')
#     plt.show()
#     plt.figure(figsize=[10, 10])
#     plt.imshow(hsv[:, :, 0], cmap='gray')
#     plt.title('h')
#     plt.show()
    plt.figure(figsize=[10, 10])
    plt.imshow(cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB))
    plt.show()

    
def hist_eq_hsv(hsv_image):
    hsv_image = hsv_image.copy()
    hsv_image[:, :, 2] = cv2.equalizeHist(hsv_image[:, :, 2])
    return hsv_image

def change_hue(hsv_image):
    hsv_image = hsv_image.copy()
    hsv_image[:, :, 0] = (hsv_image[:, :, 0] - 50) % 180
    return hsv_image
    
# Example usage
rgb_image_path = '../data/onions_2.jpg'
hsv_image_onions = convert_rgb_to_hsv(rgb_image_path)

hsv_image_mushrooms = convert_rgb_to_hsv('../data/mushrooms_1.jpg')

dark = '../data/dark.jpeg'
hsv_dark = convert_rgb_to_hsv(dark)

medium = '../data/medium.jpeg'
hsv_medium = convert_rgb_to_hsv(medium)

bright = '../data/bright.jpeg'
hsv_bright = convert_rgb_to_hsv(bright)


# Save the HSV image
# hsv_image_path = '../data'
# cv2.imwrite(hsv_image_path, hsv_image)

```

```python
plot_hsv(hsv_bright)
plot_hsv(change_hue(hsv_bright))
plot_hsv(hist_eq_hsv(hsv_bright))
```
